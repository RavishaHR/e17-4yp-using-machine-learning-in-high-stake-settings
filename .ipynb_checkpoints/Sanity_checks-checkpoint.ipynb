Feature importance

```python
# Import libraries
import os
import pandas as pd
import pickle
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
import random
```

```python
# Function to load the pickle file and return the feature importances
def load_model_and_feat(pkl_model_path):
  # Load the model
  #model = pickle.load(open(pkl_model_path, 'rb'))
  with open(pkl_model_path,"rb") as file_handle:
    model = pickle.load(file_handle)
  # Load the feature importance array
  importance = model.feature_importances_
  # Get the feature names
  feat_names = model.feature_names_in_

  return importance, feat_names


# Function to select the top k features and plot
def select_top_features_and_plot(k, importance, feat_names, title):
  # Create df and sort
  df = pd.DataFrame({"Feature": feat_names,
                  "Importance": importance})
  df_sorted = df.sort_values("Importance", ascending=False).head(k)

  # Plot
  plt.cla()
  plt.barh(df_sorted["Feature"], df_sorted['Importance'], label=title)
  plt.show()

  return df_sorted
```

## The ratio of the average of the top 1000 projects to the average of the rest of the projects is computed over six folds.

### Fold 1

```python
def get_topk_botom_ratio(test_files,test_pred ):
    fold = pd.read_csv(os.path.join(root,test_files))
    fold_pred = pd.read_csv(os.path.join(root, test_pred ))
    Fold = pd.concat([fold, fold_pred["1"]],axis=1)
    Fold = Fold.drop(["Unnamed: 0", "Project ID"],axis=1)
    Fold_sort = Fold.sort_values(["1"],ascending=False)
    ratio = Fold_sort[:1000].mean() / Fold_sort[1000:].mean()
    positive_counts = Fold_sort[:1000].where(Fold_sort[:1000].abs()>0).count()
    negative_counts = Fold_sort[1000:].where(Fold_sort[1000:].abs()>0).count()
    Fold1_ratio = pd.concat([ratio, positive_counts,negative_counts], axis=1).rename(columns={0:"ratio",1:"count_posstive",2:"count_negative"})
    return Fold1_ratio
```

```python
print(f"Test file {test_files[0]} added.")
get_topk_botom_ratio(test_files[0],test_pred[0] )
```
